```python
import pandas as pd
from pandas_datareader import data
import datetime as dt
import matplotlib.pyplot as plt
```

```python
start = dt.datetime(2019,1,1)

data.DataReader("AAPL","yahoo", "2019-01-01")
```

```python
aapl = data.DataReader("AAPL","yahoo", "2019-01-01")
aapl
```

```python
aapl.plot()
```

```python
aapl['Close'].plot()
```

```python
aapl[['Open','Close']].plot(figsize = (16,10))
```

```python
aapl[aapl['Close'] > aapl['Close'].mean()]

#returns all the closing dates that the closing price is higher thsan the mean price
```

```python
#locate rows after 2020 jan 1st
aapl['Close'].loc['2020-01-01':'2020-12-31'].plot()
```

```python
#30 day moving average
aapl["Close"].rolling(window=30).mean().plot(label = 'Close Mean')
#shows the moving average price

#then add the closing price
aapl['Close'].plot()
plt.legend()
```

```python
#closing prices for facebook and apple (you can add more with this technique)
portfolio = data.DataReader(['FB',"AAPL",'GOOGL',"AMZN", "TSLA"], "yahoo", "2019-01-01")
#save it to csv

portfolio.to_csv("FANGFund.csv")
#closing prices of each stock 
portfolio["Close"].plot(figsize = (16,10))

```

```python
portfolio.head(10)
```

```python
portfolio["Close"].idxmin()
#gives the day of the minimum closing prices 
```

```python
portfolio["Close"].diff()
#calculates the difference by day for each 



```

```python
portfolio["Close"].pct_change()
#gives percentage change per day
```

```python
portfolio["Close"].pct_change().idxmax()

#the day that the percentage change was the highest

##the day that the percentage change was the lowest

portfolio["Close"].pct_change().idxmin()


#the values they were
portfolio["Close"].pct_change().max()
portfolio["Close"].pct_change().min()

```

```python
portfolio["Close"].pct_change().corr()
```
